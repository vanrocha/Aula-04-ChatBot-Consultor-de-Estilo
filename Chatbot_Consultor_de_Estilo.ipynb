```python
# Import the Phython SDK


import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('secret_kay')
genai.configure (api_key = api_key)


import random

# Simula√ß√£o de banco de dados com informa√ß√µes de pe√ßas de roupa
roupas = {
    "casual": ["camiseta", "cal√ßa jeans", "t√™nis"],
    "formal": ["camisa social", "cal√ßa social", "sapato social"],
    "festa": ["vestido", "salto alto", "clutch"],
    "frio": ["casaco", "cachecol", "bota"],
}

# Simula√ß√£o de hist√≥rico de compras do usu√°rio (inicialmente vazio)
historico_compras = []

def obter_preferencias():
    """Simula a obten√ß√£o de prefer√™ncias do usu√°rio."""
    estilos = ["casual", "formal", "festa", "frio"]
    print("Quais estilos voc√™ gosta? (selecione os n√∫meros)")
    for i, estilo in enumerate(estilos):
        print(f"{i+1}. {estilo.capitalize()}")

    escolhas = input("Digite os n√∫meros separados por v√≠rgula: ")
    indices = [int(x.strip()) - 1 for x in escolhas.split(",")]
    estilos_preferidos = [estilos[i] for i in indices]
    return estilos_preferidos

def gerar_sugestao(estilos_preferidos):
    """Gera sugest√µes de roupas com base nos estilos preferidos."""
    if not estilos_preferidos:
        return "Voc√™ precisa escolher pelo menos um estilo para receber sugest√µes."

    estilo_escolhido = random.choice(estilos_preferidos)
    pecas = roupas[estilo_escolhido]
    sugestao = f"Que tal um look {estilo_escolhido} com: {', '.join(pecas)}?"
    return sugestao

def registrar_compra(peca):
    """Simula o registro de uma compra no hist√≥rico do usu√°rio."""
    historico_compras.append(peca)
    print(f"{peca.capitalize()} adicionado ao seu hist√≥rico de compras!")

def mostrar_historico():
    """Mostra o hist√≥rico de compras do usu√°rio."""
    if not historico_compras:
        print("Seu hist√≥rico de compras est√° vazio.")
    else:
        print("Suas compras anteriores:")
        for peca in historico_compras:
            print(f"- {peca.capitalize()}")

# Intera√ß√£o com o usu√°rio
print("Ol√°! Sou seu Consultor de Estilo. üòä")

estilos_preferidos = obter_preferencias()

while True:
    print("\nO que voc√™ gostaria de fazer?")
    print("1. Receber sugest√£o de look")
    print("2. Registrar uma compra")
    print("3. Ver meu hist√≥rico de compras")
    print("4. Sair")

    opcao = input("Digite o n√∫mero da op√ß√£o: ")

    if opcao == '1':
        sugestao = gerar_sugestao(estilos_preferidos)
        print(sugestao)
    elif opcao == '2':
        peca = input("Qual pe√ßa voc√™ comprou? ")
        registrar_compra(peca)
    elif opcao == '3':
        mostrar_historico()
    elif opcao == '4':
        print("At√© mais! üëã")
        break
    else:
        print("Op√ß√£o inv√°lida. Tente novamente.")

import random

#Melhorando a visualiza√ß√£o
#C√≥digo dispon√≠vel em https://ai.google.dev/tutorials/python_quickstart#import_packages
import textwrap
from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('‚Ä¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

#Imprimindo o hist√≥rico
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
  print('-------------------------------------------')
```
