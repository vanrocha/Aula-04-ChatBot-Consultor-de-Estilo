```python
# Import the Phython SDK


import google.generativeai as genai
from google.colab import userdata
api_key = userdata.get('secret_kay')
genai.configure (api_key = api_key)


import random

# SimulaÃ§Ã£o de banco de dados com informaÃ§Ãµes de peÃ§as de roupa
roupas = {
    "casual": ["camiseta", "calÃ§a jeans", "tÃªnis"],
    "formal": ["camisa social", "calÃ§a social", "sapato social"],
    "festa": ["vestido", "salto alto", "clutch"],
    "frio": ["casaco", "cachecol", "bota"],
}

# SimulaÃ§Ã£o de histÃ³rico de compras do usuÃ¡rio (inicialmente vazio)
historico_compras = []

def obter_preferencias():
    """Simula a obtenÃ§Ã£o de preferÃªncias do usuÃ¡rio."""
    estilos = ["casual", "formal", "festa", "frio"]
    print("Quais estilos vocÃª gosta? (selecione os nÃºmeros)")
    for i, estilo in enumerate(estilos):
        print(f"{i+1}. {estilo.capitalize()}")

    escolhas = input("Digite os nÃºmeros separados por vÃ­rgula: ")
    indices = [int(x.strip()) - 1 for x in escolhas.split(",")]
    estilos_preferidos = [estilos[i] for i in indices]
    return estilos_preferidos

def gerar_sugestao(estilos_preferidos):
    """Gera sugestÃµes de roupas com base nos estilos preferidos."""
    if not estilos_preferidos:
        return "VocÃª precisa escolher pelo menos um estilo para receber sugestÃµes."

    estilo_escolhido = random.choice(estilos_preferidos)
    pecas = roupas[estilo_escolhido]
    sugestao = f"Que tal um look {estilo_escolhido} com: {', '.join(pecas)}?"
    return sugestao

def registrar_compra(peca):
    """Simula o registro de uma compra no histÃ³rico do usuÃ¡rio."""
    historico_compras.append(peca)
    print(f"{peca.capitalize()} adicionado ao seu histÃ³rico de compras!")

def mostrar_historico():
    """Mostra o histÃ³rico de compras do usuÃ¡rio."""
    if not historico_compras:
        print("Seu histÃ³rico de compras estÃ¡ vazio.")
    else:
        print("Suas compras anteriores:")
        for peca in historico_compras:
            print(f"- {peca.capitalize()}")

# InteraÃ§Ã£o com o usuÃ¡rio
print("OlÃ¡! Sou seu Consultor de Estilo. ðŸ˜Š")

estilos_preferidos = obter_preferencias()

while True:
    print("\nO que vocÃª gostaria de fazer?")
    print("1. Receber sugestÃ£o de look")
    print("2. Registrar uma compra")
    print("3. Ver meu histÃ³rico de compras")
    print("4. Sair")

    opcao = input("Digite o nÃºmero da opÃ§Ã£o: ")

    if opcao == '1':
        sugestao = gerar_sugestao(estilos_preferidos)
        print(sugestao)
    elif opcao == '2':
        peca = input("Qual peÃ§a vocÃª comprou? ")
        registrar_compra(peca)
    elif opcao == '3':
        mostrar_historico()
    elif opcao == '4':
        print("AtÃ© mais! ðŸ‘‹")
        break
    else:
        print("OpÃ§Ã£o invÃ¡lida. Tente novamente.")

import random

#Melhorando a visualizaÃ§Ã£o
#CÃ³digo disponÃ­vel em https://ai.google.dev/tutorials/python_quickstart#import_packages
import textwrap
from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('â€¢', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))

#Imprimindo o histÃ³rico
for message in chat.history:
  display(to_markdown(f'**{message.role}**: {message.parts[0].text}'))
  print('-------------------------------------------')
```
